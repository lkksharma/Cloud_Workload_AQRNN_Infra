```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("./gwa_t_2_anon_jobs_sqlite/Grid5000.sqlite")

query="""
SELECT
    CAST((j.SubmitTime / 300) AS INTEGER) * 300 AS WindowStart,

    COUNT(j.JobID) AS TotalJobs,
    SUM(j.ReqNProcs) AS TotalCPUReq,
    SUM(j.RunTime * j.NProc) / 300.0 AS CPUUtil,
    AVG(j.WaitTime) AS AvgWait,
    AVG(j.RunTime) AS AvgRun,
    SUM(CASE WHEN j.Status = 1 THEN 1 ELSE 0 END) * 1.0 / COUNT(*) AS SuccessRatio,

    SUM(CASE WHEN ji.JobID IS NOT NULL THEN 1 ELSE 0 END) * 1.0 / COUNT(*) AS FracInteractive,
    SUM(CASE WHEN jr.JobID IS NOT NULL THEN 1 ELSE 0 END) * 1.0 / COUNT(*) AS FracReserved,

    -- Time-of-day (local hour per site, optional)
    ( ( (CAST((j.SubmitTime / 3600) AS INTEGER) % 24) ) ) AS HourOfDay

FROM Jobs j
LEFT JOIN Jobs_Interactive ji ON j.JobID = ji.JobID
LEFT JOIN Jobs_Reservation jr ON j.JobID = jr.JobID
LEFT JOIN SitesTime st ON j.OrigSiteID = st.SiteID

WHERE j.SubmitTime IS NOT NULL
  AND j.RunTime IS NOT NULL
  AND j.WaitTime IS NOT NULL
  AND j.NProc IS NOT NULL
  AND j.ReqNProcs IS NOT NULL
  AND j.RunTime >= 0
  AND j.WaitTime >= 0
  AND j.Status != -1

GROUP BY CAST((j.SubmitTime / 300) AS INTEGER)
ORDER BY WindowStart;
"""

df = pd.read_sql_query(query, conn)
conn.close()
df.to_csv("workload_timeseries_grid5000.csv", index=False)
```

```python
import pennylane as qml
from sklearn.decomposition import PCA
import os
import numpy as np

DEFAULT_SCALE_ANGLE = np.pi
DEFAULT_SHOTS_NOISY = 400
RNG = np.random.RandomState(42)
```

```python
def load_preprocesses_data(file_path = "processes_workload_data.npz"):
    data = np.load(file_path)
    return data

def angle_encode(x, input_wires, scale = DEFAULT_SCALE_ANGLE):
    for i, w in enumerate(input_wires):
        qml.RY(float(scale) * float(x[i]), wires = w)

```

```python
def _validate_topology(top):
    valid = {"ladder", "start", "all_to_all"}
    if top not in valid:
        raise ValueError(f"give correct topology name from: {valid}")
    return top

def local_rotations(params_block, wires):
    for i, w in enumerate(wires):
        ry = params_block[i, 0]
        rz = params_block[i, 1]
        qml.RY(ry, wires = w)
        qml.RZ(rz, wires = w)

def entangler_layer(wires, topology="ladder"):
    topology = _validate_topology(topology)
    if topology == "all_to_all":
        for i in range(len(wires)):
            for j in range(lwn(wires)):w
                qml.CNOT(wires=[wires[i],wires[j]])
    elif topology == "start":
        center = wires[0]
        for w in wires[1:]:
            qml.CNOT(wires = [center, w])
    else:
        for i, j in zip(wires[:-1], wires[1:]):
            qml.CNOT(wires=[i,j])

def cry_gates(input_wires, hidden_wires, angles):
    n = min(len(input_wires), len(hidden_wires))

    if pnp.ndim(angles) == 0:
        for i in range(n):
            qml.CRY(angles, wires = [input_wires[i], hidden_wires[i]])
    else:
        for i in range(n):
            qml.CRY(angles[i], wires = [input_wires[i], hidden_wires[i]])
```

```python
import jax
import jax.numpy as jnp
from jax import grad, jit, vmap

class AQRNNCell:
    def __init__(self,
                n_x: int,
                n_h: int,
                seq_len: int,
                n_layers: int = 1,
                topology: str = "ladder",
                param_sharing: bool = False,
                angle_scale: float = np.pi):

        assert n_x >=1 and n_h >= 1 and seq_len >= 1
        self.n_x = n_x
        self.n_h = n_h
        self.seq_len = seq_len
        self.n_layers = max(1, int(n_layers))
        self.topology = _validate_topology(topology)
        self.param_sharing = bool(param_sharing)
        self.angle_scale = float(angle_scale)

        self.total_qubits = n_x + n_h
        self.input_wires = list(range(0, n_x))
        self.hidden_wires = list(range(n_x, n_x + n_h))
        self.total_wires = self.input_wires + self.hidden_wires

        if self.param_sharing:
            self.param_shape = (self.n_layers, self.total_qubits, 2)
        else:
            self.param_shape = (self.seq_len, self.n_layers, self.total_qubits, 2)

    def _pqc_block(self, params_block):
        local_rotations(params_block, self.total_wires)
        entangler_layer(self.total_wires, self.topology)
        gate_angle = pnp.mean(params_block[:, 0]) * 0.1
        cry_gates(self.input_wires, self.hidden_wires, gate_angle)

    def build_qnode(self, device = None, device_mode="simulator", shots=400, noise_model=None):
        if device is None:
            if device_mode == "cuda":
                dev = qml.device("lightning.gpu", wires=self.total_qubits)
            if device_mode == "simulator":
                dev = qml.device("default.qubit", wires=self.total_qubits)
            elif device_mode == "noisy":
                dev = qml.device("qiskit.aer", wires=self.total_qubits, backend="qasm_simulator", shots=shots, noise_model=moise_model)
        else:
            dev = device
        param_shape = self.param_shape

        @qml.qnode(dev, interface="jax")
        def sequence_qnode(sequence, params, forget_gate=None):
            
            for t in range(self.seq_len):
                x_t = sequence[t]
                # 1. Load: Encode new input
                angle_encode(x_t, self.input_wires, scale = self.angle_scale)
                
                # 2. Process: Interaction/Entanglement
                #param sharing
                if self.param_sharing:
                    for layer in range(self.n_layers):
                        block_params = params[layer]
                        self._pqc_block(block_params)
                else:
                    for layer in range(self.n_layers):
                        block_params = params[t, layer]
                        self._pqc_block(block_params)
                # forget ggate
                if forget_gate is not None:
                    fg = forget_gate[t]

                    if pnp.ndim(fg) == 0:
                        angle = float((1.0 - fg) *0.25)
                        for hw in self.hidden_wires:
                            qml.RY(angle, wires=hw)
                    else:
                        for i, hw in enumerate(self.hidden_wires):
                            angle = float((1.0 - float(fg[i])) *0.25)
                            qml.RY(angle, wires=hw)
                
                # 3. Reset: Unload input to keep memory pure (Option A)
                qml.adjoint(angle_encode)(x_t, self.input_wires, scale=self.angle_scale)

            return [qml.expval(qml.PauliZ(x)) for x in self.hidden_wires]
        return sequence_qnode, param_shape, dev
```

```python
import os
import time
from pennylane import numpy as pnp
from sklearn.metrics import mean_squared_error
import pickle

class AdamState:
    def __init__(self, shape, lr=1e-2, beta1=0.9, beta2=0.999, eps=1e-8):
        self.m = np.zeros_like(shape, dtype=float)
        self.v = np.zeros_like(shape, dtype=float)
        self.beta1 = beta1
        self.beta2 = beta2
        self.lr =lr
        self.eps = eps
        self.t = 0
```

```python
def adam_update(param, grad, state: AdamState):
    state.t += 1
    state.m = state.beta1 * state.m + (1 - state.beta1) * grad
    state.v = state.beta2 * state.v + (1 - state.beta2) * (grad ** 2)

    m_hat = state.m / (1 - state.beta1 ** state.t)
    v_hat = state.v / (1 - state.beta2 ** state.t)

    step = state.lr * m_hat / (np.sqrt(v_hat) + state.eps)
    return param-step, state

    
def pack_classical_weights(W1, b1, W2, b2):
    return np.concatenate([W1.flatten(), b1.flatten(), W2.flatten(), b2.flatten()])

def unpack_classical_weights(vec, n_h, hidden_dim):
    idx = 0
    sW1 = hidden_dim* n_h
    W1 = vec[idx: idx + sW1].reshape((hidden_dim, n_h)); idx += sW1
    sb1 = hidden_dim
    b1 = vec[idx: idx + sb1].reshape((hidden_dim,)); idx += sb1
    sW2 = hidden_dim * 1
    W2 = vec[idx: idx + sW2].reshape((1, hidden_dim)); idx += sW2
    sb2 = 1
    b2 = vec[idx: idx + sb2].reshape((1,)); idx += sb2
    return W1, b1, W2, b2


# classical head

def mlp_forward(readout, wvec, n_h, hidden_dim):
    W1, b1, W2, b2 = unpack_classical_weights(wvec, n_h, hidden_dim)

    W1 = pnp.array(W1)
    b1 = pnp.array(b1)
    W2 = pnp.array(W2)
    b2 = pnp.array(b2)
    readout = pnp.array(readout)

    if readout.ndim == 1:
        h = pnp.maximum(0, pnp.dot(W1, readout) + b1)
        out = pnp.dot(W2, h) + b2
        return out[0]
    else:
        z = pnp.dot(readout, W1.T) + b1
        h = pnp.maximum(0, z)
        out = pnp.dot(h, W2.T) + b2
        return out[:, 0]
```

```python
def make_loss_fn(qnode, seq_le, n_x, n_h, hidden_dim):
    def loss(params_q, wvec, batch_X, batch_y):
        B = batch_X.shape[0]
        total_loss = 0.0

        for i in range(B):
            seq = batch_X[i]

            readouts = qnode(seq, params_q)
            readouts = pnp.array(readouts, requires_grad=False)
            pred = mlp_forward(readouts, wvec, n_h, hidden_dim)

            pred = pnp.array(pred)
            y = pnp.array(batch_y[i])
            total_loss += (pred -y) **2
        return total_loss / B
    return loss
```

```python
# Assuming numpy as np, time, pickle, and pennylane as qml are imported globally.
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from scipy.stats import pearsonr

def train_aqrnn(
    npz_path = "processed_workload_data.npz",
    seq_len=10,
    n_h = 2,
    hidden_dim = 4,
    n_layers =1,
    param_sharing = False,
    n_epochs = 2,
    batch_size = 2,
    lr_q = 1e-4,
    lr_c = 1e-4,
    device_mode = "simulator",
    shots = 100,
    device_noise_model = None,
    seed = 42,
    save_path ="aqrnn_checkpoint.pkl"):

    np.random.seed(seed)
    
    # Assuming scaler_y is loaded from the npz file or is globally available
    # e.g., if it's in the npz: scaler_y = data['scaler_y'].item()
    data = np.load(npz_path, allow_pickle=True)
    X_train = data["X_train"]
    y_train = data["y_train"].reshape(-1)
    X_val = data["X_val"]
    y_val = data["y_val"].reshape(-1)
    X_test = data["X_test"]
    y_test = data["y_test"].reshape(-1)
    print(f"Loaded data: Train {X_train.shape}, Val {X_val.shape}, Test {X_test.shape}")

    n_x = X_train.shape[2]

    cell = AQRNNCell(n_x=n_x, n_h=n_h, seq_len=seq_len, n_layers=n_layers, param_sharing=param_sharing)
    qnode, param_shape, dev = cell.build_qnode(device_mode=device_mode, shots=shots, noise_model=device_noise_model)
    print("Using device:", dev, "with parameter shape:", param_shape)
    print(f"AQRNN node built with outputs: {qnode}")

    params_q = 0.05 * np.random.randn(*param_shape).astype(float)

    W1 = 0.1* np.random.randn(hidden_dim, n_h).astype(float)
    b1 = np.zeros((hidden_dim,), dtype=float)
    W2 = 0.1* np.random.randn(1, hidden_dim).astype(float)
    b2 = np.zeros((1,), dtype=float)
    wvec = pack_classical_weights(W1, b1, W2, b2)

    q_state = AdamState(params_q, lr=lr_q)
    c_state = AdamState(wvec, lr=lr_c)

    loss_fn = make_loss_fn(qnode, seq_len, n_x, n_h, hidden_dim)

    grad_q_fn = qml.grad(loss_fn, argnum=0)
    grad_c_fn = qml.grad(loss_fn, argnum=1)

    N = X_train.shape[0]
    n_batches = int(np.ceil(N / batch_size))

    best_val_rmse = 1e9
    best_ckpt = None

    history = {"train_rmse": [], "val_rmse": []}
    
    for epoch in range(1, n_epochs + 1):
        t0 = time.time()
        print("Current runnning epoch", epoch, " at time ", t0)

        perm = np.arange(N)
        train_losses = []

        for b in range(n_batches):
            bi = b * batch_size
            bj = min((b+1)* batch_size, N)
            batch_idx = perm[bi:bj]
            batch_X = X_train[batch_idx]
            batch_y = y_train[batch_idx]

            grad_q = grad_q_fn(params_q, wvec, batch_X, batch_y)
            grad_c = grad_c_fn(params_q, wvec, batch_X, batch_y)

            params_q, q_state = adam_update(params_q, grad_q, q_state)
            wvec, c_state = adam_update(wvec, grad_c, c_state)

            batch_loss = loss_fn(params_q, wvec, batch_X, batch_y).item()
            train_losses.append(batch_loss)

        # This function uses the *current* parameters from the epoch
        def predict_all(X_arr):
            preds = []
            for i in range(X_arr.shape[0]):
                seq = X_arr[i]
                readouts = qnode(seq, params_q)
                preds.append(mlp_forward(np.array(readouts), wvec, n_h, hidden_dim))
            # --- FIX 1: Ensure output is flattened to match y_train/y_val shapes ---
            return np.array(preds).reshape(-1)
        
        train_preds = predict_all(X_train)
        val_preds = predict_all(X_val)
        train_rmse = np.sqrt(mean_squared_error(y_train, train_preds))
        val_rmse = np.sqrt(mean_squared_error(y_val, val_preds))
        history["train_rmse"].append(train_rmse)
        history["val_rmse"].append(val_rmse)

        if val_rmse < best_val_rmse:
            best_val_rmse = val_rmse
            best_ckpt = {
                "params_q": params_q,
                "wvec": wvec,
                "epoch": epoch,
                "train_rmse": train_rmse,
                "val_rmse": val_rmse
            }
            with open(save_path, "wb") as f:
                pickle.dump(best_ckpt, f)
        t1 = time.time()
        print(f"Epoch {epoch}/{n_epochs} - Train RMSE: {train_rmse:.6f}, Val RMSE: {val_rmse:.6f}, Time: {t1 - t0:.2f}s")

    print("Training complete. Best Val RMSE:", best_val_rmse)

    best_params = best_ckpt["params_q"]
    best_wvec = best_ckpt["wvec"]

    # --- FIX 2: Reuse prediction logic with best parameters ---
    # Define a clean prediction function to use the *best* parameters
    def predict_with_best_params(X_arr):
        preds = []
        for i in range(X_arr.shape[0]):
            seq = X_arr[i]
            readouts = qnode(seq, best_params) # Use best_params
            preds.append(mlp_forward(np.array(readouts), best_wvec, n_h, hidden_dim)) # Use best_wvec
        return np.array(preds).reshape(-1)

    # Replace the redundant manual loop with a single function call
    test_preds = predict_with_best_params(X_test)
    y_test_flat = y_test.reshape(-1)
    # --- End Fix 2 ---

    # core metrics
    test_mae = mean_absolute_error(y_test_flat, test_preds)
    test_mse = mean_squared_error(y_test_flat, test_preds)
    test_rmse = np.sqrt(test_mse)
    test_r2 = r2_score(y_test_flat, test_preds)

    # --- FIX 3: Removed redundant/less robust MAPE and Corr calculations ---
    # The calculations below are more robust (handle divide-by-zero)
    # and were overwriting these anyway.
    
    # nonzero_mask = y_test_flat != 0
    # test_mape = np.mean(np.abs((y_test_flat[nonzero_mask] - test_preds[nonzero_mask]) / y_test_flat[nonzero_mask])) * 100
    # test_corr, _ = pearsonr(y_test_flat, test_preds)
    # --- End Fix 3 ---

    if np.std(y_test_flat) > 0 and np.std(test_preds) > 0:
        test_corr, _ = pearsonr(y_test_flat, test_preds)
    else:
        test_corr = float("nan")

    eps = 1e-8
    denom = np.maximum(np.abs(y_test_flat), eps)
    test_mape = 100.0 * np.mean(np.abs((y_test_flat - test_preds) / denom))

    def smape_npy(a, b, eps = 1e-8):
        denom = (np.abs(a) + np.abs(b)) / 2.0
        mask = denom > eps
        if not np.any(mask):
            return float("nan")
        return 100.0 * np.mean(np.abs(a[mask] - b[mask]) / denom[mask])
    test_smape = smape_npy(y_test_flat, test_preds)

    #converting to original scale for relatable metrics (without blowing up)
    y_test_orig = scaler_y.inverse_transform(y_test_flat.reshape(-1, 1)).reshape(-1)
    test_preds_orig = scaler_y.inverse_transform(test_preds.reshape(-1, 1)).reshape(-1)

    test_mae_orig = mean_absolute_error(y_test_orig, test_preds_orig)
    test_mse_orig = mean_squared_error(y_test_orig, test_preds_orig)
    rmse_orig = np.sqrt(test_mse_orig)
    r2_orig = r2_score(y_test_orig, test_preds_orig)
    
    # --- FIX 4: Add std dev check for original correlation ---
    if np.std(y_test_orig) > 0 and np.std(test_preds_orig) > 0:
        test_corr_orig, _ = pearsonr(y_test_orig, test_preds_orig)
    else:
        test_corr_orig = float("nan")
    # --- End Fix 4 ---
        
    denom_orig = np.maximum(np.abs(y_test_orig), 1e-6)
    test_mape_orig = 100.0 * np.mean(np.abs((y_test_orig - test_preds_orig) / denom_orig))
    test_smape_orig = smape_npy(y_test_orig, test_preds_orig)


    print("\nEvaluation metrics (scaled values):")
    print(f"  MAE  : {test_mae:.6f}")
    print(f"  RMSE : {test_rmse:.6f}")
    # print(f"  R2   : {test_r2:.6f}")
    # print(f"  MAPE : {test_mape:.2f}%")
    # print(f"  Corr : {test_corr:.4f}")

    # print("\nEvaluation metrics (original units):")
    # print(f"  MAE  : {test_mae_orig:.6f}")
    # print(f"  RMSE : {rmse_orig:.6f}")
    # print(f"  R2   : {r2_orig:.6f}")
    # print(f"  MAPE : {test_mape_orig:.2f}% (denom floored to 1e-6)")
    # print(f"  SMAPE: {test_smape_orig:.6f}%")
    # print(f"  Corr : {test_corr_orig:.4f}")

    out = {
    "best_ckpt": best_ckpt,
    "history": history,
    "metrics": {
        "scaled": {
            "mae": float(test_mae),
            "rmse": float(test_rmse),
            "r2": float(test_r2),
            "mape": float(test_mape),
            "smape": float(test_smape),
            "corr": None if np.isnan(test_corr) else float(test_corr)
        },
        "original": {
            "mae": float(test_mae_orig),
            "rmse": float(rmse_orig),
            "r2": float(r2_orig),
            "mape": float(test_mape_orig),
            "smape": float(test_smape_orig),
            "corr": None if np.isnan(test_corr_orig) else float(test_corr_orig)
        }
    }
}
    with open("aqrnn_final.pkl", "wb") as f:
        pickle.dump(out, f)
    return out
```

```python
import numpy as np

data = np.load("processed_workload_data.npz")
X_train = data["X_train"]
X_val = data["X_val"]
X_test = data["X_test"]
y_train = data["y_train"]
y_val = data["y_val"]
y_test = data["y_test"]

X_all = np.concatenate([X_train, X_val, X_test])
y_all = np.concatenate([y_train, y_val, y_test])

N_keep = 1000
rng = np.random.RandomState(42)
idx = rng.choice(len(X_all), size = N_keep, replace = False)

X_subset = X_all[idx]
y_subset = y_all[idx]

n = len(X_subset)
i1 = int(0.7 * n)
i2 = int(0.85 * n)

np.savez("subset500_processed_workload_data.npz", X_train = X_subset[:i1], X_val = X_subset[i1:i2], X_test = X_subset[i2:], y_train = y_subset[:i1], y_val = y_subset[i1:i2], y_test = y_subset[i2:])

print(fr"Subset data of length {n} saved as subset500_processed_workload_data.npz")
```

```python
import numpy as np
from sklearn.decomposition import PCA
data = np.load("subset500_processed_workload_data.npz")

def reduceThenSave(data, n_components = 2):
    pca = PCA(n_components, random_state = 42)
    X_train = data["X_train"]
    N, T, F = X_train.shape
    flat = X_train.reshape(-1, F)
    pca.fit(flat)

    def apply(arr):
        N_, T_, F_ = arr.shape
        flat_ = arr.reshape(-1, F_)
        reduced = pca.transform(flat_)
        return reduced.reshape(N_, T_, n_components)
    
    X_train_reduced = apply(data["X_train"])
    X_val_reduced = apply(data["X_val"])
    X_test_reduced = apply(data["X_test"])

    np.savez("subset500_processed_workload_data.npz", X_train = X_train_reduced, X_val = X_val_reduced, X_test = X_test_reduced, y_train = data["y_train"], y_val = data["y_val"], y_test = data["y_test"])

    print(f"PCA reduced data of length {n_components} saved as subset500_processed_workload_data.npz")

reduceThenSave(data, n_components = 4)
    
```
